```python
import matplotlib.pyplot as plt
import numpy as np
from minkfncts2d import MF2D
```

```python
# Make an image of the given shape with gaussian noise centered at 0.5, sigma 0.1 and values in range [0, 1]
# Then calculate the 2D Minkowski functionals at 100 thresholds in [0, 1]
def test_gaussian(shape):
    image = np.clip(np.random.normal(0.5, 0.1, shape), 0.0, 1.0)
    F = []
    U = []
    Chi = []
    
    for threshold in np.linspace(0.0, 1.0, 100):
        (f, u, chi) = MF2D(image, threshold)
        F.append(f)
        U.append(u)
        Chi.append(chi)

    return (F, U, Chi, image)
```

```python
(F, U, Chi, image) = test_gaussian((1000, 1000))
```

```python
plt.figure(1)
plt.clf()
plt.title("Image with Gaussian Noise")
plt.imshow(image, cmap="viridis")
plt.show()
```

```python
fig = 1
for t in np.linspace(0.2, 0.8, 5):
    timage = image.copy()
    timage[timage >= t] = 1.0
    timage[timage < t] = 0.0
    plt.figure(fig)
    plt.clf()
    plt.title("Image thresholded at "+"{:.2f}".format(t))
    plt.imshow(timage, cmap="viridis")
    fig += 1
plt.show()
```

```python
plt.figure(1)
plt.clf()
x = np.linspace(0.0, 1.0, 100)
plt.plot(x, F, x, U, x, Chi)
plt.title("2D Minkowski Functions")
plt.legend(["F (Area)", "U (Boundary)", "$\chi$ (Euler characteristic)"])
plt.xlabel("Threshold")
plt.show()
```
